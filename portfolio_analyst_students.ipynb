<h2 style="color: #FF6347;">Gráfico de la rentabilidad histórica acumulada de la cartera.</h2>

```python
#Calculo del retorno diario para cada activo
retornos_diarios = porfolio_df.pct_change()

#Calculo de la rentabilidad diaria promedio de la cartera
retornos_diarios['Portfolio'] = retornos_diarios.mean(axis=1)

#Calculo de la rentabilidad acumulada del Portfolio
retornos_acumulados = (1 + retornos_diarios['Portfolio']).cumprod() - 1

#Gráfico de la rentabilidad histórica acumulada
plt.figure(figsize=(12, 6))
plt.plot(retornos_acumulados, label="Rentabilidad Acumulada del Portfolio", color="teal")
plt.title("Rentabilidad Histórica Acumulada del Portfolio")
plt.xlabel("Fecha")
plt.ylabel("Rentabilidad Acumulada")
plt.legend()
plt.grid(True)
plt.show()
```
